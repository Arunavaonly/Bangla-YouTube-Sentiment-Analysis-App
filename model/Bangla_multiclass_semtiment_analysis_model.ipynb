```python
import pandas as pd
df = pd.read_csv('/content/bangla_sentiment.csv', encoding = 'utf-8', on_bad_lines ='skip', index_col = False)
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
df['Label'].value_counts()
```

```python
df.rename(columns ={'Label':'labels'}, inplace = True)
df
```

```python
!pip install transformers datasets
```

```python
from datasets import Dataset
data = Dataset.from_pandas(df).train_test_split(test_size =0.1, shuffle = True)
print(data['train'][:5])
print(data['test'][:5])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")

def tokenize(examples):
  return tokenizer(examples['Data'], truncation = True, max_length = 512)

data = data.map(tokenize, batched = True)
print(data['train'])
print(data['test'])
```

```python
#Preparing model
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained('xlm-roberta-base', num_labels = 3)
```

```python
#creating a compute metrics function to incorporate in the trainer
from sklearn.metrics import f1_score

def compute_metrics(pred):
  labels = pred.label_ids
  preds = pred.predictions.argmax(-1)
  f1 = f1_score(labels, preds, average = 'weighted')
  return {'f1':f1}
```

```python
!pip install accelerate -U

from transformers import TrainingArguments

batch_size = 16
training_args = TrainingArguments(output_dir = 'Bangla_multiclass_sentiment_analysis_model', learning_rate= 3e-5, num_train_epochs =3,
    per_device_train_batch_size = batch_size,
    per_device_eval_batch_size= batch_size,
    eval_steps = 200,
    weight_decay = 0.01,
    evaluation_strategy= "steps",
    fp16 = True
                                  )
```

```python
from transformers import Trainer
import torch

# Check if GPU is available
if torch.cuda.is_available():
  device = "cuda:0"  # You can specify a GPU index if you have multiple GPUs
  print(f'Using GPU: {torch.cuda.get_device_name(0)}')

else:
    device = "cpu"

# Create your model and move it to the specified device
model = model.to(device)

trainer = Trainer(
    model = model,
    args = training_args,
    compute_metrics = compute_metrics,
    train_dataset = data['train'],
    eval_dataset = data['test'],
    tokenizer = tokenizer,
)
```

```python
#clearing pytorch cuda cache to avoid MemoryRunOutError
import torch
torch.cuda.empty_cache()
```

```python
trainer.train()
```

```python
from sklearn.metrics import classification_report,f1_score
def compute_metrics(pred):
  labels = pred.label_ids
  preds = pred.predictions.argmax(-1)
  f1 = f1_score(labels, preds, average = None, labels = labels)
  f1_scores_with_labels = {label:score for label,score in zip(labels, f1)}
  report = classification_report(labels, preds, target_names =['Class 0', 'Class 1', 'Class 2'])

  return {'f1':f1_scores_with_labels, 'report': report}


f1_trainer = Trainer(
    model=model,
    args=training_args,
    tokenizer = tokenizer,
    train_dataset=data['train'],
    eval_dataset=data['test'],
    compute_metrics=compute_metrics,
)
f1_trainer.evaluate()
```
